```python
#maximum sub array
lst=list(map(int,input().split()))
cs=lst[0]
ms=lst[0]
for i in range(len(lst)):
        if cs<0:
            cs=lst[i]
        else:
            cs=cs+lst[i]
        if cs>ms:
            ms=cs
print(ms)
```

```python
#give sum from subarray
nums = [-1, 2, 3, -3, 6]
curr_sum = nums[0]
max_sum = nums[0]
start = 0
end = 0
temp_start = 0
for i in range(1, len(nums)):

    if nums[i] > curr_sum + nums[i]:
        curr_sum = nums[i]
        temp_start = i
    else:
        curr_sum = curr_sum + nums[i]
    if curr_sum > max_sum:
        max_sum = curr_sum
        start = temp_start
        end = i

print(max_sum)
print(nums[start:end+1])
```

## Two pointer /SLIDING WINDOW


```python
#give sum from the array
lst=list(map(int,input().split()))
gs=int(input())
left=0
right=len(lst)-1
while left<right:
    sum=lst[left]+lst[right]
    if sum==gs:
       print(lst[left],lst[right])
       break
    elif sum<gs:
        left+=1
    else:
        right-=1
print(sum)
```
