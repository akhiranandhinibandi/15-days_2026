```python

# armstrong number
n=int(input())
temp=n
sum=0
while temp>0:
       digit=temp%10
       sum+=digit**3
       temp//=10
if n==sum:
    print("yes")

else:
    print("no")
```

```python

#palindrome or not
n=int(input())
temp=n
rev=0
while temp>0:
    digit=temp%10
    rev=rev*10+digit
    temp//=10
if n==rev:
    print("yes")
else:
    print("no")
```

```python
#frequency of an element
n=int(input())
lst=list(map(int,input().split()))
res=[0]*n
for i in range(n):
  if(res[i]==1):
    continue
  count=1
  for j in range(i+1,n):
    if lst[i]==lst[j]:
      count+=1
      res[j]=1
  print(lst[i],count)
```

```python
#first repeating element in a list
n=int(input())
lst=list(map(int,input().split()))
for i in range(n):
  for j in range(i+1,n):
    if(lst[i]==lst[j]):
      print(lst[i])
      flag=1
      break
  break
if(flag==0):
  print("No repeating element")
```
